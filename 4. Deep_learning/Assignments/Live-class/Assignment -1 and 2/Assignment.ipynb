## Question 4

### Learning rate:

Learning rate decides how fast the model will converge towards the minimum error. In neural networks, we employ the gradient descent method to converge to local minima as we want to minimize the difference between actual output and the predicted output. Therefore, learning rate determines the rate of convergence. We need to choose optimum value that ensures we will reach 
minima with least EPOCHS. If it is too large, then we can end up missing minima as the jump will be too high. If too small,it will take a lot of time to converge or won't even converge as the weight update won't occur.

#### Case_1: With learning rate

```python
x_1 = 2
x_2 = 5
y_act = 31
```

```python
lr = 0.001
w_1 = 3
w_2 = 7

for epoch in range(60):
    y_pre = w_1*x_1 + w_2*x_2
    err = (y_act - y_pre)**2
    dEw1 = 2*(y_act - y_pre)*(-x_1)
    dEw2 = 2*(y_act - y_pre)*(-x_2)

    w_1 = w_1 - lr*dEw1
    w_2 = w_2 - lr*dEw2
    print("\nValue of w1 - ", w_1, "\nValue of w2 - ", w_2, "\nError is - ", err)
```

```python
print(w_1, w_2) #best weights
```

#### Case_2: Without learning rate

```python
#lr = 0.001
w_1 = 3
w_2 = 7

for epoch in range(40):
    y_pre = w_1*x_1 + w_2*x_2
    err = (y_act - y_pre)**2
    dEw1 = 2*(y_act - y_pre)*(-x_1)
    dEw2 = 2*(y_act - y_pre)*(-x_2)

    w_1 = w_1 - dEw1  # Here, we have removed learning rate
    w_2 = w_2 - dEw2  # Here, we have removed learning rate
    print("\nValue of w1 - ", w_1, "\nValue of w2 - ", w_2, "\n Error is - ", err)
```

So, as we can see that from the above 2 cases that once we remove the learning rate, the error does not reduce and instead it increases without the learning. It is only when we introduce the learning rate, then the error starts reducing to the minimum. In other words, weights start updating during back-propagation phase and finally we can see convergence towards minimum in the gradient descent algorithm.

## Question-5

There are 3 phases in the deep learning projects and they are:

1. Forward Pass
2. Error Calculation
3. Back propagation


## An example use-case:

Let's say that we want to use a simple neural network in which we have the input x, with some weight w and bias b. Let's say we have been given the actual output y for given value of x. So, we need to predict the value of y and see how close actual value is to the predicted value. This we can do using the 3 step life cycle of deep learning as defined below:



## 1. Forward pass phase

The step performed below is forward pass, In forward pass, we pass the data points through a particular perceptron, multiplying the inputs with weights and adding bias. Finally, we get output as predicted values as well as the error.
 Suppose, we have the input as x and bias as the b. Then, the output variable/dependent variable is given by the equation
 
$$ y = wx + b $$
                                                 
Where, w is the weight associated with the input, and b is the bias. y will be the predicted value that we will get. 

## 2. Error Calculation

After performing forward pass, we get the predicted output. However, we can see that there is a difference between actual output and the predicted output. This is called as error. We need to reduce this error. The error is given by the equation below:

$$ Error = (Yactual - Ypredicted)^2 $$
 
Where, Y_actual is the actual output and Y_predicted is the predicted output.


## 3. Back Propagation Phase

As we know, we need to reduce the error so that the predicted output becomes equal to the actual output. So, the mechanism we use for it is known as the back propagation. In back propagation, once we get the predicted output, we take the corresponding weights and then we use the gradient descent algorithm to find out the best weights by minimizing the error. The best weights are the weights which give us the predicted value closest to the actual value.

$$ dE/dw(i) = d/dw(i)((Yactual - Ypredicted)^2) $$

Finally, after differentiating, we get 

$$ dE/dw(i) = 2*(Yactual - Ypredicted)*(-x(i)) $$

Now, updating the weights is given by the formula

$$ w(i) = w(i) - learning rate*dE/dw(i) $$

where, i goes from 1 to the defined number of EPOCHS.

Finally, we get the value of y which is closest to the actual y.

### Part a.

```python
x1 = [10, 8, 15]
x2 = [8, 20, 25]
y = [30, 15, 20] #actual output

n = len(y) #The interger specifying length if y(output)
```

## 1. Forward pass phase

The step performed below is forward pass, In forward pass, we pass the data points through a particular perceptron, multiplying the inputs with weights and adding bias. Finally, we get output as predicted values as well as the error.
 Suppose, we have the input as x and bias as the b. Then, the output variable/dependent variable is given by the equation
 
$$ y = wx + b $$
                                                 
Where, w is the weight associated with the input, and b is the bias.

For multiple inputs, the equation will be
                                                
$$ y = w1*x1 + w2*x2 + w3*x3 + b $$

## 2. Error Calculation

After performing forward pass, we get the predicted output. However, we can see that there is a difference between actual output and the predicted output. This is called as error. We need to reduce this error. The error is given by the equation below:

$$ Error = (Yactual - Ypredicted)^2 $$
 
Where, Y_actual is the actual output and Y_predicted is the predicted output.



```python
#Feed-forward 

def forward_pass(w1, w2, b):

    total_error = 0
    y_pred = []
    
    for i in range(n):  #Here, please note that it is not epoch because it is just forward pass but epoch invoves both forward and backward pass

        y_hat = x1[i]*w1 + x2[i]*w2 + b   #y_hat = x1.w1+x2.w2+b, it is the predicted output, the one our network predicts
        E = (y[i] - y_hat)**2    #Calculating the error
        y_pred.append(y_hat)
        total_error+=E      #total error of all points
    return total_error, y_pred
    
```

```python
total_error, y_pred  = forward_pass(10, 5, -10)
print(total_error)
```

## 3. Back Propagation Phase

As we know, we need to reduce the error so that the predicted output is equal to the actual output. So, the mechanism we use for it is known as the back propagation. In back propagation, once we get the predicted output, we take the corresponding weight and then we use the gradient descent algorithm to find out the best weights by minimizing the error. The best weights are the weights which give us the predicted value closest to the actual value.

$$ dE/dw(i) = d/dw(i)((Yactual - Ypredicted)^2) $$

Finally, after differentiating, we get 

$$ dE/dw(i) = 2*(Yactual - Ypredicted)*(-x(i)) $$

Now, updating the weights is given by the formula

$$ w(i) = w(i) - learning rate*dE/dw(i) $$

where, i goes from 1 to the defined number of EPOCHS.

```python
##Back-propagation algorithm

def weightUpdate(w1, w2, b):
    for i in range(n):
        dE_dw1 = -2 * x1[i] * (y[i] - y_pred[i])
        dE_dw2 = -2 * x2[i] * (y[i] - y_pred[i])
        dE_db = -2 * (y[i] - y_pred[i])
        
        w1 -= lr * dE_dw1
        w2 -= lr * dE_dw2
        b -= lr * dE_db
    return w1, w2, b
```

```python
# Random initialization of weights

random_init = np.random.randn(3) 
w1, w2, b = random_init[0], random_init[1], random_init[2]
lr = 0.0001

error = list()
for i in range(400):
    total_error, y_pred = forward_pass(w1,w2,b)
    error.append(total_error)
    print(f"Epoch: {i} | MSE = {round(total_error, 2)}")
    w1, w2, b = weightUpdate(w1, w2, b)
```

```python
import matplotlib.pyplot as plt

plt.plot(error)
plt.xlabel("EPOCH")
plt.ylabel("ERROR")
plt.grid()
plt.show()
```

## a. Sigmoid function

```python
def sigmoid(x):
    
    z = 1/(1 + np.exp(-x))
    return z
```

```python
x = np.linspace(-10, 10, 100)
```

```python
z = sigmoid(x)
```

```python
plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("Sigmoid(X)")
  
plt.show()
```

## tanh(x)

```python
def tanhx(x):
    
    j = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
    return j
```

```python
j = tanhx(x)
```

```python
plt.plot(x,j)
plt.xlabel("x")
plt.ylabel("Tanh(X)")
  
plt.show()
```

## ReLU

```python
def ReLU(x):
    
    k = np.maximum(0, x)
    return k
```

```python
k = ReLU(x)
```

```python
plt.plot(x,k)
plt.xlabel("x")
plt.ylabel("Tanh(X)")
  
plt.show()
```

## SoftMax

```python
def softmax(x):
    l = np.exp(x)/sum(np.exp(x))
    return l
```

```python
l = softmax(x)
```
