### PROGRAMMING SINGLE LAYER PERCEPTRON

```python
#Inputs to the perceptron, so it will be like first 3 will go for x1, 10 for x2 and 17 for x3, then 7 for x1, 16 for x2 and 16 for x3......
x1 = np.random.randint(3, 12, 10)
x2 = np.random.randint(9, 18, 10)
x3 = np.random.randint(12, 20, 10)
y = x1*7 + 5*x2 + 4*x3
```

```python
x1,x2,x3
```

```python
# defining actual y(output)

y
```

```python
SHAPE = x1.shape[0]
SHAPE
```

```python
y = y.reshape(SHAPE,1) #reshaping y
y
```

```python
X = np.array([x1, x2, x3])
X
```

```python
#Here, first row signifies first values of x1,x2,x3 and so on...
X = X.T
X
```

```python
X.shape
```

### Now, defining weights

Weights are nothing but they simply signify that how much importance each feature has got... 
More, the weight, more the associated input has an effect on output

```python
#Randomly generating weights

w = np.array([np.random.randn(1),
              np.random.randn(1),
              np.random.randn(1)])
w
```

### Defining learning rate

Learning rate decided how fast the model will converge towards the minimum. In neural networks, we employ the gradient descent method to converge to local minima as we want to minimize the difference between actual output and the predicted output. Therefore, learning rate determines the rate of convergence. We need to choose optimum such value that ensures we will reach 
 minima. If it is too large, then we can end up missing minima. If too small,it will take a lot of time to converge or won't
 even converge as the weight update won't occur.

```python
lr = 0.0001
```

```python
### Here we are running 50 EPOCHS. The one EPOCH signifies both forward and backward propagation of an entire dataset in a single pass.
```

```python
error_list = []
for i in range(50):
    #forward pass
    y_pred = np.matmul(X, w)
    error = (y - y_pred)**2
    
    #backward pass
    dEw1 = 2*np.matmul((y - y_pred).T, (-X[:,0].reshape(SHAPE, 1)))
    dEw2 = 2*np.matmul((y - y_pred).T, (-X[:,1].reshape(SHAPE, 1)))
    dEw3 = 2*np.matmul((y - y_pred).T, (-X[:,2].reshape(SHAPE, 1)))
    
    w[0][0] = w[0][0] - lr * dEw1  #updating the weights for the back propagation algorithm
    w[1][0] = w[1][0] - lr * dEw2
    w[2][0] = w[2][0] - lr * dEw3
    
    error_list.append(error.mean())
    print("value of w1 - ", w[0][0], "value of w2 - ", w[1][0], "value of w3 - ", w[2][0])
    print(error)
```

```python
plt.plot(error_list)
```
